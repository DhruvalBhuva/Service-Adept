```python
import numpy as np 
import pandas as pd
```

```python
df = pd.read_csv('Pincode_30052019.csv')
df
```

```python
df_test = df[['Office Name', 'District', 'StateName']][(df['District'] == 'BHAVNAGAR') | (df['District'] == 'RAJKOT' ) | (df['District'] == 'NASHIK' ) ].copy()
df_test
```

```python
df_test['Office Name'] = df_test['Office Name'].str.replace('BO', '')
df_test['Office Name'] = df_test['Office Name'].str.replace('B.O', '')
df_test['Office Name'] = df_test['Office Name'].str.replace('S.O', '')
df_test['Office Name'] = df_test['Office Name'].str.replace('SO', '')
```

```python
df_test.columns = ['area', 'city', 'state']
```

```python
df_city = df_test[['city', 'state']].copy()
```

```python
df_city = df_city.drop_duplicates()
df_city = df_city.reset_index(drop=True)
df_city = df_city.reset_index()
```

```python
df_city['index'] += 1
```

```python
df_city.to_csv('cities.csv', columns = ['index', 'city', 'state'], index = False)
```

```python
df_city = df_city.set_index('city')
```

```python
df_city_dict = df_city['index'].to_dict()
```

```python
df_city_dict
```

```python
df_test['city'] = df_test['city'].replace(df_city_dict)
```
